```python
from skimage import feature
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#---------- Read Image ----------#

img = mpimg.imread('Downloads/AX .jpg')

print (type(img))
print (img.shape, img.dtype)
print (img[100,200],img[100,200],img[100,200],img[100,200])
print (img.max(),img.min())

M = np.zeros((img.shape[0],img.shape[1]))
print (M)

M[:,:] = img[:,:]

print (M.max(),M.min(),M.shape)

plt.imshow(M, cmap = plt.get_cmap('gray'))

plt.title("ax T2 caipi")
plt.savefig("ax T2 caipi.jpg")

edges = feature.canny(M, sigma=3.5)

fig, ax = plt.subplots()

ax.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
#ax.axis('off')
ax.set_title('Canny Edge Detection')

plt.savefig("ax T2 caipi.jpg")
#plt.show()
```

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
im = Image.open('Downloads/S5.jpg')
channels = im.split()
z = np.array(channels[0])
N  = z.size
n_classes = int(N**.4)
fig = plt.figure()
plt.clf()
fig.add_subplot(2, 1, 1)
plt.imshow(z, origin = "upper")
plt.colorbar()
fig.add_subplot(2, 1, 2)
plt.title('Histogramme')
plt.ylabel('Nombre de pixels')
plt.xlabel('Valeurs des pixels')
plt.hist(z.flatten(), bins=n_classes, histtype = "stepfilled")
plt.grid()
plt.show()
```

```python
from skimage import feature
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('Downloads/AX SPACE T2 Caipi D.jpg')


print (img.shape, img.dtype)
print (img[100,200],img[100,200],img[100,200],img[100,200])
print (img.max(),img.min())

M = np.zeros((img.shape[0],img.shape[1]))
print (M)

M[:,:] = img[:,:]

print (M.max(),M.min(),M.shape)

plt.imshow(M, cmap = plt.get_cmap('gray')

edges = feature.canny(M, sigma=2.5)

fig, ax = plt.subplots()

ax.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
#ax.axis('off')
ax.set_title('Canny Edge Detection')

plt.savefig("ax T2 caipi.jpg")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Downloads/AX T2 HOMO.jpg')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(122),plt.imshow(dst)
plt.xticks([]),plt.yticks([])
plt.savefig("ax T2 caipi filtr.jpg")
plt.show()
```
